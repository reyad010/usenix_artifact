```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
import re

def extract_latencies(file_path):
    latencies = []
    with open(file_path, 'r') as file:
        for line in file:
            match = re.search(r'CHA \d+, Addr \d+: (\d+)', line)
            if match:
                latencies.append(int(match.group(1)))
    return latencies
# Example usage
file_path = "../output/lat/s1r_s2r_s0c_64stride.txt"
s1r_s2r_s0c_64stride = extract_latencies(file_path)
# Example usage
file_path = "../output/lat/s1r_s2r_s0c_65536stride.txt"
s1r_s2r_s0c_65536stride = extract_latencies(file_path)

```

```python
# plot histogram of the data
fig, ax = plt.subplots()
ax.hist(s1r_s2r_s0c_64stride, bins=1000, alpha=0.5, label='Hitme miss')
ax.hist(s1r_s2r_s0c_65536stride, bins=1000, alpha=0.5, label='s1r_s2r_s0c_65536stride')
plt.xlim(0, 800)
plt.xlabel('Latency (Cycles)')
plt.ylabel('Frequency')
ax.legend()
```

```python
#  Plot the CDF of the data
fig, ax = plt.subplots()
ax.hist(s1r_s2r_s0c_64stride, bins=100, density=True, histtype='step', cumulative=True, label='Hitme miss')
ax.hist(s1r_s2r_s0c_65536stride, bins=100, density=True, histtype='step', cumulative=True, label='Hitme hit')
plt.xlim(0, 1000)
plt.ylim(0, 1)
plt.xlabel('Latency (Cycles)')
plt.ylabel('CDF')
ax.legend()
```
